```python
# Example 1: point plot
# https://matplotlib.org/

from matplotlib import pyplot

x = [i for i in range(0, 51, 10)]
y = [i**2 for i in x] # y = f(x)

pyplot.plot(x,y, marker = 's')

pyplot.show()
```

```python
# Example 2: point plot with description
# https://numpy.org/
# uncomment the third plt.plot(...) method to see random point plot

import numpy as np
import matplotlib.pyplot as plt
import random

random.seed()
x = np.arange(0, 4*np.pi, 0.1) # change setup
y1 = np.sin(x)
y2 = np.sin(2*x)/4+1
y3 = [random.uniform(-1, 1)*1.2 for _ in x]

fig, ax = plt.subplots(layout = 'constrained')

ax.plot(x, y1, color = 'green', lw = 1.5, marker = 'x', ms = 3, label = '$f_1(x)$')
ax.plot(x, y2, color = 'k', lw = 3, label = '$f_2(x)$')
plt.plot(x, y3, color = 'r', lw = 1, label = '$f_3(x)$')

ax.set_ylabel('y(x)')
ax.set_xlabel('x')
ax.set_title('Example function y(x)')
ax.legend(loc='upper left', ncols = 2)
ax.set_ylim(-1.5, 2)

plt.show()
```

```python
# Example 3: CDF
import matplotlib.pyplot as plt
#import pandas as pd
import numpy as np
import random

fig, ax = plt.subplots(layout = 'constrained')

# raw data
x = [random.uniform(0, 100) for _ in range(20)]

# manually computed CDF
cdf_x = np.sort(x)
cdf_y = (np.arange(len(x))+1)/ len(x)
ax.plot(cdf_x, cdf_y, marker = 'x', label = 'manual')

# one-line solution in Matplotlib
ax.ecdf(x, label = 'ECDF()')

ax.set_ylim(0, 1.05)
ax.legend()
ax.set_ylabel('Probability')
ax.set_title('Example CDF')
plt.show()
```
